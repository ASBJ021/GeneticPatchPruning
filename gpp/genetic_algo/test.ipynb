# GA Score Histograms

Quick histograms for confidence and feature preservation scores.

```python
import json
import matplotlib.pyplot as plt

# Update this path to your JSONL log if needed
# log_path = '/home/utn/firi22ka/Desktop/jenga/GeneticPatchPruning/logs/fitness_logs_3_cosine_sim/fitness_scores.jsonl'
# log_path = '/home/utn/firi22ka/Desktop/jenga/GeneticPatchPruning/logs/new_fitness_logs_cosine_cs/fitness_scores_cls_cosine.jsonl'
# log_path = '/home/utn/firi22ka/Desktop/jenga/GeneticPatchPruning/logs/new_fitness_logs_normalized_cs/fitness_scores_cls_logits_normalized.jsonl'
log_path = '/home/utn/firi22ka/Desktop/jenga/GeneticPatchPruning/logs/new_fitness_logs_normalized_cs/fitness_scores_masked_cls_normalized.jsonl'

records = []
with open(log_path, 'r', encoding='utf-8') as f:
    for line in f:
        line = line.strip()
        if not line:
            continue
        records.append(json.loads(line))

conf = [r['confidence_score'] for r in records]
feat = [r['feature_preservation_score'] for r in records]

fig, axes = plt.subplots(1, 2, figsize=(12, 4), tight_layout=True)
axes[0].hist(conf, bins=30, color='#4C72B0', alpha=0.8)
axes[0].set_title('Confidence Score (normalized maskedcls)')
axes[0].set_xlabel('score')
axes[0].set_ylabel('count')

axes[1].hist(feat, bins=30, color='#DD8452', alpha=0.8)
axes[1].set_title('Feature Preservation Score')
axes[1].set_xlabel('score')
axes[1].set_ylabel('count')

plt.show()
```

## Visualizing Patch Selector 


```python
import json
import matplotlib.pyplot as plt
import numpy as np

path = "/home/utn/firi22ka/Desktop/jenga/GeneticPatchPruning/Model_0.3/PatchSelector/imagenet-1k_0_0.3_PatchSelector_TS_cliplabels_full_merged/benchmark_results/per_img_metrics.jsonl"
xs, ys = [], []
with open(path, 'r', encoding='utf-8') as f:
    for line in f:
        line = line.strip()
        if not line:
            continue
        r = json.loads(line)
        if 'keep_tokens' in r and 'inf_time_per_img' in r:
            xs.append(r['keep_tokens'])
            ys.append(r['inf_time_per_img'])

print(f"Loaded {len(xs)} points")
plt.figure(figsize=(12, 8))
plt.scatter(xs, ys, s=10, alpha=0.5)
plt.xlabel('keep_tokens')
plt.ylabel('inf_time_per_img (s)')
plt.title('Inference Time vs Keep Tokens')
plt.grid(True, alpha=0.3)

# set x-axis tick step to 5
x_min, x_max = min(xs), max(xs)
plt.xticks(np.arange(x_min, x_max + 1, 2))

plt.tight_layout()
plt.show()
```

## GA Distill: First 100 Image IDs

Plots for `mem_usage_image_encoder`, `keep_tokens`, and `image_id` from the first 100 records.

```python
import json
import matplotlib.pyplot as plt

path = "/home/utn/firi22ka/Desktop/jenga/GeneticPatchPruning/Model_0.3_ga_distill/mlp_0.5_1.0/PS_ga_distill_fit_0.8_cls_1.2/per_img_metrics_adaptive_topk_full.jsonl"
limit = 100
rows = []
with open(path, 'r', encoding='utf-8') as f:
    for line in f:
        line = line.strip()
        if not line:
            continue
        record = json.loads(line)
        if {'image_id', 'mem_usage_image_encoder', 'keep_tokens'}.issubset(record):
            rows.append(record)

rows = sorted(rows, key=lambda r: r['image_id'])[:limit]

image_ids = [r['image_id'] for r in rows]
memory_usage = [r['mem_usage_image_encoder'] for r in rows]
keep_tokens = [r['keep_tokens'] for r in rows]

print(f"Loaded {len(rows)} records for image_id range {image_ids[0]} to {image_ids[-1]}")

fig, axes = plt.subplots(3, 1, figsize=(12, 16))

axes[0].plot(image_ids, memory_usage, marker='o', markersize=4, linewidth=1)
axes[0].set_title('Memory Usage vs Image ID (First 100)')
axes[0].set_xlabel('image_id')
axes[0].set_ylabel('mem_usage_image_encoder (MB)')
axes[0].grid(True, alpha=0.3)

axes[1].scatter(keep_tokens, memory_usage, s=24, alpha=0.7)
axes[1].set_title('Memory Usage vs Keep Tokens (First 100)')
axes[1].set_xlabel('keep_tokens')
axes[1].set_ylabel('mem_usage_image_encoder (MB)')
axes[1].grid(True, alpha=0.3)

axes[2].plot(image_ids, keep_tokens, marker='o', markersize=4, linewidth=1)
axes[2].set_title('Keep Tokens vs Image ID (First 100)')
axes[2].set_xlabel('image_id')
axes[2].set_ylabel('keep_tokens')
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python

mem_min = min(memory_usage)
mem_max = max(memory_usage)
pad = 0.01 * (mem_max - mem_min if mem_max > mem_min else 1.0)

keep_tokens_min = min(keep_tokens) if keep_tokens else 0
keep_tokens_max = max(keep_tokens) if keep_tokens else 196
keep_tokens_pad = 0.01 * (keep_tokens_max - keep_tokens_min if keep_tokens_max > keep_tokens_min else 1.0)

inf_time = [r['inf_time_per_img'] for r in rows]
inf_time_min = min(inf_time) if inf_time else 0
inf_time_max = max(inf_time) if inf_time else 1



plt.figure(figsize=(12, 8))
# plt.plot(image_ids, memory_usage, marker='o', markersize=4, linewidth=1)
plt.scatter(image_ids, memory_usage, s=24, alpha=0.7)
plt.title('Memory Usage vs Image ID')
plt.xlabel('image_id')
plt.ylabel('mem_usage_image_encoder (MB)')
plt.ylim(4.7, 5)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 8))
plt.scatter(keep_tokens, memory_usage, s=24, alpha=0.7)
plt.title('Memory Usage vs No. of Selected Patches')
plt.xlabel('No of Selected Patches')
plt.ylabel('mem_usage_image_encoder (MB)')
plt.ylim(4.7, 5)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 8))
# plt.plot(image_ids, keep_tokens, marker='o', markersize=4, linewidth=1)
plt.scatter(image_ids,keep_tokens, s=24, alpha=0.7)
plt.title('Selected Patches vs Image ID')
plt.xlabel('image_id')
plt.ylabel('No of Selected Patches')
plt.ylim(90, 105)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 8))
plt.scatter(keep_tokens, inf_time, s=24, alpha=0.7)
plt.title('Inference Time vs No. of Selected Patches')
plt.xlabel('No of Selected Patches')
plt.ylabel('inf_time_per_img (s)')
plt.ylim(0.005, 0.009)
plt.grid(True, alpha=0.3)   

plt.tight_layout()
plt.show()
```
