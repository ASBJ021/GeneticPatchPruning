```python
import json
import os
from typing import List
import matplotlib.pyplot as plt
import numpy as np
```

```python
scores_path = "/home/utn/firi22ka/Desktop/jenga/GeneticPatchPruning/logs/fitness_logs/fitness_scores.jsonl"
records = []

if os.path.exists(scores_path):
    with open(scores_path, "r", encoding="utf-8") as f:
        for line in f:
            try:
                record = json.loads(line)
                records.append(record)
                # print(f'{record =  }')
            except Exception:
                print(Exception)
                continue
else:
    print('No path found ')
```

```python
records[0]
```

```python
conf = [r['confidence_score'] for r in records]
feat = [r['feature_preservation_score'] for r in records]
score = [r['fitness_score'] for r in records]
```

```python
conf[0]
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4), tight_layout=True)

axes[0].hist(conf, bins=30, color="#4C72B0", alpha=0.8)
axes[0].set_title("Confidence Score")
axes[0].set_xlabel("score")
axes[0].set_ylabel("count")

axes[1].hist(feat, bins=30, color="#DD8452", alpha=0.8)
axes[1].set_title("Feature Preservation Score")
axes[1].set_xlabel("score")
axes[1].set_ylabel("count")

plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.hist(conf, bins=30, alpha=0.6, label="confidence")
plt.hist(feat, bins=30, alpha=0.6, label="feature_preservation")
plt.legend()
plt.xlabel("score")
plt.ylabel("count")
plt.title("Score Distributions")
plt.show()
```
